1. Visualize the survival rate of passengers based on their ticket class (1st, 2nd, or 3rd class).
Were the passengers in higher classes more likely to survive?

```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
survival_by_class = titanic.groupby('class')['survived'].mean().reset_index()
survival_by_class
```

Explanation here:

```python
sns.barplot(x='class', y='survived', data=titanic)
```

The passengers in first class till third class were most likely to survive.
